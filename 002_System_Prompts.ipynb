### System prompt

```python
import boto3
client=boto3.client("bedrock-runtime",region_name="ap-south-1")
model_id="apac.anthropic.claude-3-haiku-20240307-v1:0"
messages=[]
def user_message(messages,text):
    message={
    "role":"user",
    "content":[{"text":text}]
    }
    messages.append(message)

def assisstant_message(messages,text):
     message={
    "role":"assistant",
    "content":[{"text":text}]
    }
     messages.append(message)

def chat(messages,temperature=1.0,stop_sequences=[]):
     system_prompt='''You are a gym trainer that helps with workout queries, gym split optimizations, and supplement suggestions. Only answer queries related to this'''
     response=client.converse(modelId=model_id,messages=messages,system=[{"text":system_prompt}],inferenceConfig={"temperature":temperature,"stopSequences":stop_sequences})
     messages.append(response["output"]["message"])
     print(response["output"]["message"]["content"][0]["text"])

user_message(messages,"Is it better to host a backend server using AWS EC2 or EKS?")
chat(messages)
```

### Temperature

```python
messages=[]
user_message(messages,"give the best gym split in a sentence")
chat(messages,temperature=0.0)
```

### Streaming

```python
messages=[]
user_message(messages,"give an explanation not less than 100 words on why protein is necessary for fat loss")
response=client.converse_stream(messages=messages,modelId=model_id)
for event in response["stream"]:
    if "contentBlockDelta" in event:
        chunk=event["contentBlockDelta"]["delta"]["text"]
        print(chunk,end="")
```
