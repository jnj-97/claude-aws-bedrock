### Multi-turn Conversations

```python
messages=[]
def user_message(messages,text):
    message={
    "role":"user",
    "content":[{"text":text}]
    }
    messages.append(message)

def assisstant_message(messages,text):
     message={
    "role":"assistant",
    "content":[{"text":text}]
    }
     messages.append(message)

def chat(messages):
     response=client.converse(modelId=model_id,messages=messages)
     messages.append(response["output"]["message"])
     print(response["output"]["message"]["content"][0]["text"])

user_message(messages,"whats your model name")
chat(messages)
user_message(messages,"Can you eloborate further")
chat(messages)
messages
```

### chatbot/conversation

```python
def infinite_chat():
    message_array=[]
    while True:
        user_text=input("> ")
        user_message(message_array,user_text)
        chat(message_array)

infinite_chat()
```
